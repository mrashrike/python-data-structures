### Fibonacci Series

```python
n = 5 

a: int = 0
b: int = 1

for i in range(0, n):
    c = a + b
    a = b
    b = c
    # Fibonaci starts with 1 not with 0, n=5 means index starts from 1,2,3,4,5
  
```
